```python
def encode_features(df):
    features = ['workclass', 'marital-status','occupation','race', 'sex','native-country', 'income']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])    
    return df

adult_df = encode_features(adult_df)
adult_df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

def format_features(df):
    features = ['workclass', 'marital-status','occupation','race', 'sex','native-country', 'income']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df


def transform_features(df):
    df = format_features(df)
    return df

y_adult_df = adult_df['income']
X_adult_df = adult_df.drop('income',axis=1)
# X_adult_df = adult_df.drop(, axis = 1)
X_adult_df = transform_features(X_adult_df)
X_adult_df = transform_features(X_adult_df)
```
